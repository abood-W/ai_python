```python
import pandas as pd
```

```python
df = pd.read_csv('/content/onlinefoods.csv')
```

```python
df.columns.values
```

```python
#Data cleaning
df.drop('Unnamed: 12', axis=1, inplace=True)
df['Gender'] = df['Gender'].map({'Female': 0, 'Male': 1})
df['Marital Status'] =df['Marital Status'].map({'Single':0, 'Married':1,'Prefer not to say':2})
df['Occupation'] = df['Occupation'].str.strip()
df['Occupation'] = df['Occupation'].map({'Student':0, 'Employee':1,'Self Employeed':2 , 'Other':3})
df['Monthly Income'] =df['Monthly Income'].map({'No Income':0 ,'Below Rs.10000':1 ,'More than 50000':2, '10001 to 25000':3,
 '25001 to 50000':4})
df['Educational Qualifications'] =df['Educational Qualifications'].map({'Post Graduate':0, 'Graduate':1, 'Ph.D':2, 'Uneducated':3 ,'School':4})
df['Output'] = df['Output'].map({'No':0, 'Yes':1})
df['Feedback'] = df['Feedback'].str.strip()
df['Feedback'] = df['Feedback'].map({'Positive':1, 'Negative':0})
df_cleaned = df.dropna()
```
