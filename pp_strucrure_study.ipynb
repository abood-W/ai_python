```python
from paddleocr import PPStructureV3

pipeline = PPStructureV3(
    use_chart_recognition=False,
    use_formula_recognition=False
    )
print("PPStructureV3 ready ")
```

```python
out = pipeline.predict("/content/PMC4760359_00006.jpg")
for res in out:
    res.print()
    res.save_to_json("output/")
    res.save_to_markdown("output/")
```

Markdown


```python
markdown_list = []
markdown_images = []

for res in out:
  md_info = res.markdown
  markdown_list.append(md_info)
  markdown_images.append(md_info.get("markdown_images", {}))
```

```python
markdown_texts = pipeline.concatenate_markdown_pages(markdown_list)
```

```python
from pathlib import Path
input_file= "/content/PMC4760359_00006.jpg"
output_path = Path("./outputttt")
mkd_file_path = output_path / f"{Path(input_file).stem}.md"
mkd_file_path.parent.mkdir(parents=True, exist_ok=True)
```

```python
with open(mkd_file_path, "w", encoding="utf-8") as f:
    f.write(markdown_texts)
```

```python
for item in markdown_images:
    if item:
        for path, image in item.items():
            file_path = output_path / path
            file_path.parent.mkdir(parents=True, exist_ok=True)
            image.save(file_path)
```

```python
import json
```

JSON

```python
import numpy as np

output_path = Path("./output_json")
output_path.mkdir(parents=True, exist_ok=True)

def to_serializable(obj):
    if isinstance(obj, (np.float32, np.float64)):
        return float(obj)
    if isinstance(obj, (np.int32, np.int64)):
        return int(obj)
    return str(obj)  # fallback

all_pages = []

for i, res in enumerate(out):
    page_info = {
        "page": i + 1,
        "texts": res.get("overall_ocr_res", {}).get("rec_texts", []),
        "tables": [tbl["pred_html"] for tbl in res.get("table_res_list", [])],
        "layout": res.get("layout_det_res", {}).get("boxes", [])
    }
    all_pages.append(page_info)

with open(output_path / "document.json", "w", encoding="utf-8") as f:
    json.dump(all_pages, f, ensure_ascii=False, indent=2, default=to_serializable)

```

TEXT

```python
output_path = Path("./output_txt")
output_path.mkdir(parents=True, exist_ok=True)

all_texts = []

for res in out:
    texts = res.get("overall_ocr_res", {}).get("rec_texts", [])
    all_texts.extend(texts)

with open(output_path / "document.txt", "w", encoding="utf-8") as f:
    f.write("\n".join(all_texts))
```

HTML

```python
output_path = Path("./output_html")
output_path.mkdir(parents=True, exist_ok=True)

for i, res in enumerate(out):
    for j, table in enumerate(res.get("table_res_list", [])):
        html = table["pred_html"]
        with open(output_path / f"page{i+1}_table{j+1}.html", "w", encoding="utf-8") as f:
            f.write(html)
```
