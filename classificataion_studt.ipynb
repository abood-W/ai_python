```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
iris = load_iris()
iris
```

```python
x = iris.data
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_predection = model.predict(X_test)
print("الدقة:", accuracy_score(y_test, y_predection))
print("تقرير التصنيف:\n", classification_report(y_test, y_predection))
```

```python
new_sample = [[5.1, 3.5, 1.4, 0.2]]  # قيم لأبعاد زهرة جديدة
prediction = model.predict(new_sample)
print("التصنيف المتوقع:", iris.target_names[prediction[0]])
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
data = load_breast_cancer()
x = data.data
y = data.target
data
```

```python
x_train , x_test,y_train , y_test = train_test_split(x,y,test_size=0.2 , random_state=42)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
predection = model.predict(x_test)
print("predection: ",accuracy_score(y_test,predection))
print(classification_report(y_test,predection))
```
